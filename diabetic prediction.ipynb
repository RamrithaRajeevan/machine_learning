```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
l=['gender', 'age', 'hypertension', 'heart_disease', 'smoking_history',
       'bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes']
for i in l:
    data[i]=le.fit_transform(data[i])
data
```

```python
x=data.drop("diabetes",axis=1)
x
```

```python
y=data["diabetes"]
y
```

```python
import seaborn as sns
sns.pairplot(data)
plt.show()
```

```python
data.plot("hypertension","diabetes",c="red")
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
from sklearn.tree import DecisionTreeClassifier
clss=DecisionTreeClassifier(criterion="entropy")
```

```python
clss.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
d1=clss.predict(xtest)
d1
```

```python
a1=accuracy_score(ytest,d1)*100
a1
```

```python
d2=clss.predict(xtrain)
d2
```

```python
a2=accuracy_score(ytrain,d2)*100
a2
```

```python
v=[[0,101,0,1,4,1334,11,6]]
w=clss.predict(v)
w
```

```python
if w==0:
    print("Not diabetic")
else:
    print("Diabetes")
```

```python
from sklearn import tree
plt.figure(figsize=(30,25))
tree.plot_tree(clss,feature_names=x.columns.tolist(),filled=True,max_depth=18)
plt.show()
```
